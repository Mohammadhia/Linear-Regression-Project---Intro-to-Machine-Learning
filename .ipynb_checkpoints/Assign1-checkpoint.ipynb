# Overview

The objective of this assignment is to implement Linear Regression on a set of data to determine its accuracy. A feature matrix is used (X) in which the inputs to the linear model are found, and predictions (Y) are made by taking the dot product of the input matrix and a weight vector (w). The predictions (Y) are compared against the actual values (T) that the model is trying to predict to see its accuracy, using a summation of the root mean squares between each y and t.

There are multiple ways of finding the optimal weight vector that will reduce the summation of the root mean square error of a linear model. The first is the Least Squares approach, in which the error of a weight vector for a given set of data will be reduced if the linear model results in a summation of the least squares. The second is the Least Mean Squares approach, in which the error of a weight vector will be reduced if the linear model results in the least AVERAGE squares.

The least mean squares iteratively modifies the weight vector through the means of gradient descent, in which the current weight has a value that it's subtracted by (involving an alpha variable and its product with the outer product of: [(weight @ x) - t and x]). This assignment uses the least mean squares linear regression model in terms of 1 step (in which the modification of the weight is done only once; online training) and in terms of iterating until the optimal weight is reached (batch training).

# Method

The Linear Model implementation is separated into two classes using the **LinearModel** class as a super class. This super class has an initializer method that automatically sets the weight vector equal to **None**, which can be used in the child classes when objects are initialized. It also has an add_ones method to add a bias column of ones to a matrix (like the X matrix that will be used to train each model). This ones column is used to accommodate for the bias variable of the weight vector that will be used. Finally, there is a "check matrix" function that will check to see if a vector is a matrix of 2 dimensions (which will be needed in the add ones function. Additionally, there are 2 abstract methods called "train" and "use" that can be implemented as needed for the LS model and the LMS models.

Descriptions of the implemented child classes can be viewed below, right before each class.

```python
import numpy as np
import abc

class LinearModel:
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abc.abstractmethod
    def train(self, X, T):
        pass
    
    @abc.abstractmethod
    def use(self, X):
        pass 
```

- **Linear Regression Model (LS Model)**
    - This class is an implementation of the LinearModel parent class. It begins with an initializer method that uses the LinearModel's default of initially setting the weight vector to **None**. After that, there is a train function that takes in a matrix of the features (X) that will be used in the regression model and a matrix of the target labels that the model will attempt to predict (T). The matrix X will have a column of ones added to it using the "add ones" function to account for bias from the weights (transforms into X1), will be reshaped from a vector to a matrix (if needed), and will finally be used in the following computation for determining the optimal weight, w, that will result in the minimum error:
        - w = inverse(X1.Transpose @ X1) @ X1.Transpose @ T
        - Please note that T will also be reshaped from a vector to a matrix (if needed) before being used in the equation above.
    - Finally, the "use" function takes the X matrix (also adding a column of ones to it to account for bias; transforming it into X1), and takes the dot product of X1 and w. The resulting predicted y is returned.

```python
# Linear Regression Class for least squares
# TODO: Finish this class for least squares
class LinearRegress(LinearModel): 
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        LinearModel.__init__(self)
        
    # train lease-squares model
    def train(self, X, T):
        """
            train linear regression model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """
        X1 = LinearModel.add_ones(self, X) #This allows for a column vector of ones to be added to account for a bias (y-intercept) in the weight vector
        X1 = X1.reshape(X1.shape[0], X1.shape[1]) #Converts from vector to matrix (if needed; if already a matrix, keeps it the same)
        T = T.reshape(T.shape[0], T.shape[1]) #Converts from vector to matrix (if needed; if already a matrix, keeps it the same)
        self.w = np.linalg.inv(X1.T @ X1) @ X1.T @ T
    
    # apply the learned model to data X
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """
        #Use the model (dot product of X and w) to predict y

        
        X1 = LinearModel.add_ones(self, X) #This allows for a column vector of ones to be added to account for a bias (y-intercept) in the weight vector
        predicted_y = X1 @ self.w
        return predicted_y
```

- **Least Mean Squares Model (LMS)**
    - This class is an implementation of the LinearModel parent class. It begins with an initializer method that uses the LinearModel's default of initially setting the weight vector to None and sets the alpha (the step value) to the alpha used as a parametr. After that, there are 2 train functions: the train_step function (for online training), and the train function (for batch learning). The main formula to find the weight that is used in the LMS model, however, must be found by iterating through each instance in the training set and is as follows:
        - **w = w - alpha * (outerproduct((w @ x1 - t), x1))**
        - The train_step function uses 1 instance of the training data (x; a 1d vector) as well as 1 instance of the expected corresponding output (t). The x has a 1 inserted at the beginning to account for bias from the weight vector (transformed to x1). The function takes the current weight and subtracts the product of alpha and the outer product of (w.Transpose @ x1 - t) and x1.
        - The train function iterates through the rows of the X matrix and uses the train_step function each time to get the optimal, overall, weight.
    - The use function then takes the dot product of X1 (the transformed X matrix with the column of ones added to account for bias) and w and returns the resulting predicted y.

```python
# LMS class 
# TODO: finish this class for LMS
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha
        
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        #Modify w for each iteration (online training)
        if(self.w is None):
            self.w = np.zeros(len(x)+1) #Initializes w to vector of 0s (Adding 1 more 0 to the vector to account for bias (y-intercept))
        
        x1 = np.insert(x, 0, 1) #Add a 1 at the beginning of the 1d vector to account for bias from first element of weight vector
        
        
        self.w = self.w - (self.alpha *   (np.outer((self.w.T @ x1) - t,  x1).reshape(self.w.shape)) ) #Can handle T of 2 dimensions because outer product is reshaped to dimensions of self.w (keeps it the same if unnecessary)
    
    # batch training by using train_step function
    def train(self, X, T):
        #batch training
        
        N = X.shape[0]
        for k in range(N):
            LMS.train_step(self, X[k], T[k])
            
    # apply the current model to data X
    def use(self, X):
        #Use the model that we got from the appropriate train function(dot product of X and w) to predict y
        
        X1 = LinearModel.add_ones(self, X) #This allows for a column vector of ones to be added to account for a bias (y-intercept) in the weight vector
        self.w = self.w.reshape(len(self.w), 1) #Converts weight from vector to matrix
        predicted_y = X1 @ self.w
        return predicted_y
```

# Results
Below are the results of the 2 methods being implemented using the iris training data. The LS method resulted in an error of ~2.319 whereas the LMS regresssion (ALL) resulted in an error of ~6.443. Of course, the LMS regression (1-step) resulted in a much higher error of ~15.349. The LMS models were both trained using an alpha of 0.02, just like in the preliminary test. Comparing the LS to the LMS (ALL) results, the LS was significantly lower. This probably indicates that the alpha used for the LMS was so high that it resulted in the error-minimizing weight being missed when gradient descent occurred. The LMS (1-step) was probably so high because it only occurred in 1 step; to lower it, the alpha would need to be increased to make this 1 step more effective.

The most important feature of the model appeared to be the petal length, as it appeared to correspond the most to the petal width in terms of distribution, as discussed above. The other features like the sepal width and length didn't appear to have that much of an impact on the model. In fact, they may have caused it to be slightly more inaccurate.
- The graphs showing the effects of the weights on the models (and the effects of the alphas on the weights) can be seen below.
- The consequences of removing the sepal width and length features can be seen further below that.

```python
def rmse(T, Y):
    return np.sqrt(np.sum((T-Y)**2))

leastSquareModel = LinearRegress()
leastSquareModel.train(iris_X, iris_T)
LS_Error = rmse(iris_T, leastSquareModel.use(iris_X))
print("Error of LS regression on Data: ", LS_Error)

leastMeanSquaresModel_All = LMS(0.02)
leastMeanSquaresModel_All.train(iris_X, iris_T)
LMS_AllError = rmse(iris_T, leastMeanSquaresModel_All.use(iris_X))
print("Error of LMS regression (all) on Data: ", LMS_AllError)

leastMeanSquaresModel_1Step = LMS(0.02)
leastMeanSquaresModel_1Step.train_step(iris_X[0], iris_T[0])
LMS_1StepError = rmse(iris_T, leastMeanSquaresModel_1Step.use(iris_X))
print("Error of LMS regression (1 step) on Data: ", LMS_1StepError)
```

```python
plt.plot(leastSquareModel.use(iris_X), 'r-', label = "LS Model Predictions")
plt.ylabel("Petal Width (cm)")
plt.title("Linear Model Predictions")
plt.legend()

plt.plot(leastMeanSquaresModel_All.use(iris_X), 'g-', label = "LMS Model (ALL) Predictions")
plt.ylabel("Petal Width (cm)")
plt.legend()

plt.plot(leastMeanSquaresModel_1Step.use(iris_X), 'y-', label = "LMS Model (1 Step) Predictions")
plt.plot(iris_T, 'b.', label = "Actual Petal Width")
plt.ylabel("Petal Width (cm)")
plt.legend()

plt.show()
```
